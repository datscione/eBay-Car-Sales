```python
autos_c.rename(columns={"yearOfRegistration":"registration_year", "monthOfRegistration":"registration_month",
                        "notRepairedDamage":"unrepaired_damage", "dateCreated":"ad_created"}, inplace=True)
```

```python
def clean_column(col):
    res = [col[0].lower()]
    for chr in col[1:]:
        if chr in ('ABCDEFGHIJKLMNOPQRSTUVWXYZ'):
            res.append('_')
            res.append(chr.lower())
        else:
            res.append(chr)
    return ''.join(res)

new_col = []
for c in autos_c.columns:
    cleaned_auto_c = clean_column(c)
    new_col.append(cleaned_auto_c)
autos_c.columns = new_col
print(autos_c.columns)
```

Let us determine brands that represent more than 5% of the vehicles.

```python
brands = autos_c["brand"].value_counts(normalize=True) # assigns
#the brand percentage value counts to variable named brands
top_brands = brands[brands>0.05].index #assigns brands with >5%
#value counts to variable named top_brands
print(top_brands)
```

Now we will determine the mean price of each of these top 5% vehicles.

```python
brand_mean_price = {}

for brand in top_brands:
    selected_brand = autos_c[autos_c["brand"] == brand]
    price_mean = selected_brand['price_USD'].mean()
    brand_mean_price[brand] = int(price_mean)
print(brand_mean_price)  
```

```python
brand_mean_price = {}
brand_mean_mileage = {}

for brand in top_brands:
    selected_brand = autos_c[autos_c["brand"] == brand]
    price_mean = selected_brand['price_USD'].mean()
    mileage_mean = selected_brand['odometer_km'].mean()
    brand_mean_price[brand] = int(price_mean)
    brand_mean_mileage[brand] = int(mileage_mean)
    bmp_series = pd.Series(brand_mean_price)
    bmm_series = pd.Series(brand_mean_mileage)
    df = pd.DataFrame(bmp_series, columns=["Mean_Price"])
    df.insert(1, "Mean_Mileage", bmm_series)
print("Brand Mean Price", brand_mean_price)  
print("Brand Mean Mileage", brand_mean_mileage)
print("Brand Mean Price Series Constructor", bmp_series)
print("Brand Mean Mileage Series Constructor", bmm_series)
print(df)
```

The mileage for all the vehicles were similar, approximately 130,000km. The car prices ranged from 3751 USD for a Ford to 9363 USD for an Audi. The price for an Audi was nearly 2.5 times that of a Ford.
